```python
df3=pd.read_csv("Products.csv")
df3
```

```python
df3.info()
```

```python
duplicate_count = df3.duplicated().sum()
print("Number of duplicates:", duplicate_count)
df3["Unit Cost USD"]=df3["Unit Cost USD"].str.replace("$","")
df3["Unit Cost USD"]=df3["Unit Cost USD"].str.replace(",","").astype(float)
df3["Unit Price USD"]=df3["Unit Price USD"].str.replace("$","")
df3["Unit Price USD"]=df3["Unit Price USD"].str.replace(",","").astype(float)
df3.info()
df3.head()
```

```python
df3.to_csv("Cleaned_Products.csv",index=False)
```

```python
df4=pd.read_csv("Sales.csv")
df4
```

```python
df4.info()
```

```python
Today=pd.to_datetime('today')
Today=Today.strftime("%Y-%d-%m")
```

```python
df4['Order Date'] = pd.to_datetime(df4['Order Date'], format='%m/%d/%Y', errors='coerce')
df4['Delivery Date'] = pd.to_datetime(df4['Delivery Date'], format='%m/%d/%Y', errors='coerce')
df4['Delivery Date'].fillna(Today, inplace=True)
duplicates = df4.duplicated().sum()
print("Number of duplicates:",duplicates)
df4.info()
df4.head(10)
```

```python
df4.to_csv("Cleaned_Sales.csv",index=False)
```

```python
# Table Creation
my_cursor.execute('''CREATE TABLE IF NOT EXISTS customers(
                  Customer_key INT PRIMARY KEY,
                  Gender VARCHAR(255),
                  Name VARCHAR(255),
                  City VARCHAR(255),
                  State_Code VARCHAR(255),
                  State VARCHAR(255),
                  Zip_Code VARCHAR(255),
                  Country VARCHAR(255),
                  Continent VARCHAR(255),
                  Birthday DATE,
                  Age INT
                  )''')
print("Table Created successfully")
```
